```python
import os
from huggingface_hub import hf_hub_download
```

```python
repo_id = "valory/autocast"
output_dir = "./data"

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

filenames = [
    "olas_benchmark.json",
    "olas_docs.pkl",
]


for filename in filenames:
    hf_hub_download(
        repo_id=repo_id,
        filename=filename,
        local_dir=output_dir,
        repo_type="dataset"
    )
```

```python
import json
import pickle
import pandas as pd
```

```python
with open(os.path.join(output_dir, "olas_benchmark.json"), "r") as f:
    data = json.load(f)

with open(os.path.join(output_dir, "olas_docs.pkl"), "rb") as f:
    docs = pickle.load(f)
```
