```python
import os
from huggingface_hub import hf_hub_download

from IPython.display import display, HTML, Markdown
```

```python
repo_id = "valory/autocast"
output_dir = "../data/autocast"

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

filenames = [
    "autocast_questions_filtered.json",
    "autocast_questions_filtered.pkl",
]
```

```python
for filename in filenames:
    hf_hub_download(
        repo_id=repo_id,
        filename=filename,
        local_dir=output_dir,
        repo_type="dataset"
    )
```

```python
import json
import pickle
import pandas as pd
```

```python
with open(os.path.join(output_dir, "autocast_questions_filtered_v3.json"), "r") as f:
    data = json.load(f)

with open(os.path.join(output_dir, "autocast_questions_filtered_v3.pkl"), "rb") as f:
    docs = pickle.load(f)
```
